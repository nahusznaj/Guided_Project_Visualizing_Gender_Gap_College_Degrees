# Guided Project: Visualizing The Gender Gap In College Degrees

This short project from [dataquest.io]() aims at practising doing plots with Matplotlib.

The dataset shows the rate of women in some degrees.

With the debatable assumption that all people that are not women are men, you can calculate the percentage of men by doing 100 - rate_women. (I know this is not the best hypothesis but I will assume so to keep it simple. In this exercise I don't intend to draw conclusions on gender itself but only on a short problem for Matplotlib.)

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

I will use colour-blind friendly colours from [Color Blind 10 palette](http://tableaufriction.blogspot.com/2012/11/finally-you-can-use-tableau-data-colors.html). So let's define dark blue and orange for women and men.

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

## Comparing more degrees in blocks

The exercise requires to use a subplot grid layout of 6 rows by 3 columns. 
And to group the degrees into STEM, liberal arts, and other, in the following way:

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 
             'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 
                 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 
              'Education', 'Agriculture','Business', 'Architecture']
```

The focus is to enable the viewer "to quickly get a high level understanding of which degrees are prone to gender imbalance and how has that changed over time".

So I do not need a lot of detail/information on the axis but just a rough idea of how gender changes in time for each degree.



```python
fig = plt.figure(figsize=(12,18))

for sp in range(0,6):
    ax = fig.add_subplot(6,3,1 + sp * 3)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",
                  labelbottom='off')
   
    if sp == 0:
        ax.text(2006, 7, 'Men')
        ax.text(2004, 87, 'Women')
    elif sp == 5:
        ax.text(2006, 90, 'Men')
        ax.text(2004, 10, 'Women')
        ax.tick_params(labelbottom='on')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)        

for sp in range(0,5):
    ax = fig.add_subplot(6,3,2 + sp * 3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]],
            c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]],
            c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",
                  labelbottom='off')
    if sp == 4:
        ax.tick_params(labelbottom='on')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)

        
for sp in range(0,6):
    ax = fig.add_subplot(6,3,3 + sp * 3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]],
            c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], 
            c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",
                   labelbottom='off')
    
    if sp == 0:
        ax.text(2006, 97, 'Women')
        ax.text(2004, 8, 'Men')
    elif sp == 5:
        ax.tick_params(labelbottom='on')
        ax.text(2006, 62, 'Men')
        ax.text(2004, 35, 'Women')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)

plt.savefig('gender_degrees.png')
```

These plots are clean, readable, uncluttered, although you can't see the detail of when exactly trends changed. As mentioned above, that's not the purpose of this visualisation.
